```python
import numpy as np
from io import BytesIO
from base64 import b64encode
from IPython.display import SVG, display

import matplotlib.pyplot as plt

# Parameters for the bell curve
mu = 0
sigma = 1
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
y = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu)/sigma)**2)

# Create the plot
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, y, label='Distribution')
sigma_v1 = -1.5
sigma_v2 = 0.2
sigma_v3 = 1.2


# Add vertical lines for Prompt versions
ax.axvline(mu + sigma_v1, color='green', linestyle='--', label=r'$\mu$(Prompt=1)')
ax.axvline(mu + sigma_v2, color='red', linestyle='--', label=r'$\mu$(Prompt=2)')
ax.axvline(mu + sigma_v3, color='purple', linestyle='--', label=r'$\mu$(Prompt=3)')



ax.set_xlabel('Z-score')
ax.set_ylabel('Probability Density')
ax.set_title('Distribution (Time = 0, Party = SPD, Question = 1)')
ax.legend()

# Remove axes for a cleaner SVG
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Save to SVG in memory
svg_buffer = BytesIO()
plt.savefig(svg_buffer, format='svg', bbox_inches='tight', transparent=True)
plt.close(fig)
svg_buffer.seek(0)
svg_data = svg_buffer.getvalue().decode()

# Display SVG in Jupyter
display(SVG(svg_data))

with open("figures/bell_curve.svg", "w") as f:
    f.write(svg_data)
```
